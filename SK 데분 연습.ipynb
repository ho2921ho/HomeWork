```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

```python
# scatter plot
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips)
```

```python
sns.catplot(x = 'day', y = 'total_bill', hue = 'smoker',col = 'sex', data = tips, split = True,kind = 'violin')
sns.catplot(x = 'day', y = 'tip', hue = 'smoker',col = 'sex', data = tips, kind = 'box')

plt.show()
```

```python
plt.pie(x =tips.smoker.value_counts(), autopct='%.0f%%', explode= [0,0.1],labels = tips.smoker.value_counts().index,colors = ['skyblue','gray'])
plt.show()
```

```python
sns.distplot(tips.total_bill, kde = True)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips,height=3.5, aspect=2)
```

```python
heat_df = pd.pivot_table(data = tips, index = 'sex', columns = 'smoker', aggfunc = 'size')

sns.heatmap(heat_df, annot = True)
```
